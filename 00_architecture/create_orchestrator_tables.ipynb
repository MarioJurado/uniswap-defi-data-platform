# This notebook creates the basic structure for both configuration and log tables.

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, BooleanType, TimestampType
import datetime
```

# Configuration table

```python
config_schema = StructType([
    StructField("layer", StringType(), False),
    StructField("load_pattern", IntegerType(), False),
    StructField("active_for_load", BooleanType(), False),
    StructField("entity_name", StringType(), False),
    StructField("notebook_path", StringType(), False)
])
```

```python
config_data = [
    ("bronze", 1, True, "factory", "../01_bronze/factory_extr"),
    ("bronze", 1, True, "pools", "../01_bronze/pools_extr"),
    ("bronze", 1, True, "swaps", "../01_bronze/swaps_extr"),
    ("bronze", 1, True, "ticks", "../01_bronze/ticks_extr"),
    ("bronze", 1, True, "tokens", "../01_bronze/tokens_extr"),
    ("silver", 1, True, "pools", "../02_silver/pools")
]
```

```python
config_table_df = spark.createDataFrame(config_data, config_schema)
config_table_df.write.format("delta").mode("overwrite").saveAsTable("uniswap.orchestrator.configuration")
```

# Logs table

```python
logs_schema = StructType([
    StructField("log_code", IntegerType(), False),
    StructField("status", StringType(), False),
    StructField("entity_name", StringType(), False),
    StructField("timestamp", TimestampType(), False),
    StructField("from_layer", StringType(), False),
    StructField("to_layer", StringType(), False),
    StructField("load_pattern", IntegerType(), False)
])
```

```python
logs_data = [
    (0, "TEST", "test_entity", datetime.datetime.now(), "bronze", "silver", 1)
]
```

```python
logs_table_df = spark.createDataFrame(logs_data, logs_schema)
logs_table_df.write.format("delta").mode("overwrite").saveAsTable("uniswap.orchestrator.logs")
```

# This notebook creates the basic structure for both configuration and log tables.

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, BooleanType, TimestampType
```

```python
config_table_path = dbutils.widgets.get("config_table_path")
logs_table_path = dbutils.widgets.get("logs_table_path")
```

# Configuration table

```python
config_schema = StructType([
    StructField("layer", StringType(), False),
    StructField("entity_name", StringType(), False),
    StructField("active_for_load", BooleanType(), False),
    StructField("priority", IntegerType(), False),
    StructField("load_pattern", IntegerType(), False),
    StructField("notebook_path", StringType(), False)
])
```

```python
config_table_df = spark.createDataFrame([], config_schema)
config_table_df.write.format("delta")\
    .option("overwriteSchema", "true")\
    .mode("overwrite")\
    .saveAsTable(config_table_path)
```

# Logs table

```python
logs_schema = StructType([
    StructField("log_code", IntegerType(), False),
    StructField("status", StringType(), False),
    StructField("layer", StringType(), False),
    StructField("entity_name", StringType(), False),
    StructField("load_pattern", IntegerType(), False),
    StructField("message", StringType(), False),
    StructField("log_timestamp", TimestampType(), False)
])
```

```python
logs_table_df = spark.createDataFrame([], logs_schema)
logs_table_df.write.format("delta")\
    .option("overwriteSchema", "true")\
    .mode("overwrite")\
    .saveAsTable(logs_table_path)
```

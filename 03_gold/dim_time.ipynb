# Entity ***Time***
+ Layer **gold**
+ Priority **1**
+ One time creation

```sql
CREATE OR REPLACE TABLE uniswap.gold.dim_time AS (
    WITH dates AS (
    SELECT
        EXPLODE(
            SEQUENCE(
                DATE('2020-01-01'),
                DATE('2030-12-31'),
                INTERVAL 1 DAY
            )
        ) AS full_date
)
SELECT
    CAST(DATE_FORMAT(full_date, 'yyyyMMdd') AS INT) AS date_key,
    full_date,
    YEAR(full_date) AS year,
    QUARTER(full_date) AS quarter,
    MONTH(full_date) AS month,
    CONCAT(
        YEAR(full_date),
        '-',
        LPAD(MONTH(full_date), 2, '0')
    ) AS year_month,
    MONTHNAME(full_date) AS month_name,
    WEEKOFYEAR(full_date) AS week_of_year,
    DAY(full_date) AS day_of_month,
    DAYOFWEEK(full_date) AS day_of_week,
    CASE
        WHEN DAYOFWEEK(full_date) = 1 THEN 7
        ELSE DAYOFWEEK(full_date) - 1
    END AS day_of_week_iso,
    DATE_FORMAT(full_date, 'EEEE') AS day_name,
    CASE
        WHEN DAYOFWEEK(full_date) IN (1,7)
        THEN TRUE
        ELSE FALSE
    END AS is_weekend,
    CURRENT_TIMESTAMP() AS _load_timestamp_gold
FROM dates
)
```
